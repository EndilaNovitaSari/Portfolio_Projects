#### Task 9 : Correlation

```python
k = 14 #number of variables for heatmap
cols = df.corr().nlargest(k, 'genre')['genre'].index
cm = df[cols].corr()
plt.figure(figsize=(15,10))
sns.heatmap(cm, annot=True, cmap = 'viridis') #call heatmap
plt.show()
```

```python
cols
```

```python
plt.figure(figsize=(15, 10))

sns.heatmap(cm[(cm >= 0.5) | (cm <= -0.2)],  #menampilkan range heatmap
            cmap='viridis', vmax=1.0, vmin=-1.0, linewidths=0.1,
            annot=True, square=True);
```
